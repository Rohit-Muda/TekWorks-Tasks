```python
df["employee_id"]=df["employee_id"].astype(str)
df["employee_id"]=df["employee_id"].str.strip()
df["employee_id"]=df["employee_id"].str.upper()
df["employee_id"]=df["employee_id"].str.replace("_","",regex=False)
df
```

# 3

```python
df["department"]=df["department"].str.strip()
df["department"]=df["department"].str.upper()
```

```python
df["salary"] = df["salary"].astype(str).str.replace("₹", "", regex=False)
df["salary"] = df["salary"].str.replace(",", "", regex=False)
df["salary"] = pd.to_numeric(df["salary"], errors="coerce")
```

```python
avg_sal=df.groupby("department")["salary"].mean()
avg_sal
```

# 12

```python
invalid_mask = df['salary'].isna() | (df['salary'] <= 0)

result = (
    df.assign(invalid=invalid_mask)
      .groupby('department')['invalid']
      .mean()          # gives fraction of invalid values
      .loc[lambda s: s > 0.25]   # keep > 25%
)
result
```
